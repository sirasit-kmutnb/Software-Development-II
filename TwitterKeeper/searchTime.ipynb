```python
import tweepy
import os
from datetime import datetime
import time
from Twitter_keeper import PullTweetsData
from Twitter_Analyzer import main
from dateutil import tz
import pytz
from threading import Thread

api_key = "SnxucIPt1fg7UUyVOT0T5j0pR"
api_key_secret = "yaToQPv95OA1fiNTHD8drKM8g8rZGM7jSQnPOLoxU3QA9UpaLm"
access_token = "1722424471-Xb0DjPVOqXsLj2sXEYXmU2sqxaDC4B793erGO6J"
access_token_secret = "D2LyXN11zAoZB0M476eb1ZGDM55oRvy4tBWNb8pR4CO0h"

auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth,wait_on_rate_limit=True)

start_time = datetime(2023, 3, 18, 0, 0, 0)  # The start time you want to retrieve tweets from
end_time = datetime(2023, 3, 20, 0, 0, 0)  # The end time you want to retrieve tweets until

query = "#ALevel66"  # The keyword you want to search for

ptw = PullTweetsData()
anl = main()

results = anl.load_sample_tweets(keyword=query)
sorted_list = sorted(results, key=lambda x: x['tweet_create_at'],reverse=True)

start_time = ptw.utc_to_local(ptw.local_to_utc(start_time))
end_time = ptw.utc_to_local(ptw.local_to_utc(end_time))

def pullbyTime(query,start_time,end_time):
    print(f"You has been called pullTweets with {query} {start_time} {end_time}")
    thread = Thread(target=pullbyTimeThread, args=(query,start_time,end_time))
    thread.start()

print(sorted_list)

def pullbyTimeThread(query,start_time,end_time):
    while True:
        searched_tweets = []
        cursor = tweepy.Cursor(api.search_tweets, q=query, lang="en", tweet_mode='extended').items()
        try:
            tweet = cursor.next()
            if ptw.utc_to_local(tweet.created_at) > start_time:
                entity_hashtag = tweet.entities.get('hashtags')
                hashtag = ptw.getHashtag(entity_hashtag)
                tweet_author = tweet.user.screen_name
                keyword = query
                dt_str = str(tweet.created_at)
                format = "%Y-%m-%d %H:%M:%S%z"
                dt_utc = datetime.strptime(dt_str, format)
                local_zone = tz.tzlocal()
                dt_local = dt_utc.astimezone(local_zone)
                tweet_create_at = dt_local
                tweet_location = tweet.user.location
                try:
                    text = tweet.retweeted_status.full_text
                except:
                    text = tweet.full_text
                tweet_post = ptw.createDictData(
                        tweet_author, tweet_create_at, tweet_location, hashtag, keyword, text)
                saveTweet = Thread(target=anl.pull_tweets.saveTweetsDict,
                                args=(tweet_post,))
                saveTweet.start()
                searched_tweets.append(tweet_post)
            elif ptw.utc_to_local(tweet.created_at) < start_time:
                break
        except tweepy.TooManyRequests:
            # time.sleep(60 * 15)
            # continue
            print(f"Rate limit exceeded. Tweet {query} has pulled until {tweet_create_at}")
            break
        except StopIteration:
            break
try:
    recentTime = ptw.utc_to_local(sorted_list[0]['tweet_create_at'])
except:
    recentTime = "No data, Pulling new tweets."
print(recentTime,start_time)
try:
    isStartTimeReached = recentTime > start_time
except:
    isStartTimeReached = False
print(isStartTimeReached)
if len(sorted_list) == 0:
    anl.pull_tweets.pullTweets(query,1000)
elif isStartTimeReached:
    pullbyTime(query,start_time,recentTime)
```
